# Pandas 시각화

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.family'] = 'Malgun Gothic'

plt.rcParams.update({
                    'text.color': "gray",
                    'axes.labelcolor': "white",
                    'xtick.color': "white",     # xtick color
                    'ytick.color': "white"      # ytick color
                    })
```

### 1. Line graph

```python
np.random.seed(2022)
df = pd.DataFrame(np.random.randn(100, 3),                      # 3그룹으로 나누어서 100개씩 뽑겠다.
                    index = pd.date_range('2022-1-1', periods=100),
                    columns=['A', 'B', 'C']).cumsum()           # 누적합
df.head()
```

```python
df.plot()
plt.title('Pandas의 plot method 사용예시')
plt.xlabel('일자'), plt.ylabel('데이터')
plt.show()          # text color는 gray
```

### 2. Bar chart

```python
import seaborn as sns
iris = sns.load_dataset('iris')

plt.figure(figsize=(12, 6))
iris.sepal_length[:20].plot(kind='barh')      # Series.plot(kind=그래프종류, rot=tick돌리기)
plt.title('Setosa 품종의 Sepal_length')
plt.show()
```

```python
df2 = iris.groupby('species').mean()
df2
```

```python
df2.plot.bar(rot=0)
plt.title('Iris 각 품종의 Feature별 평균')
plt.show()
```

```python
df2.columns.name = 'Feature'    # 범례 제목
df2.plot.bar(rot=0)
plt.title('Iris 각 품종의 Feature별 평균')
plt.ylim(0,9)       # 막대기 길이 줄이기
plt.show()
```

### 3. Pie chart

```python
titanic = sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct = '%.1f%%')
plt.title('선실별 승객수')
plt.show()
```

### 4. Box plot

```python
iris[iris.species == 'setosa'].plot.box(figsize = (6,8))
plt.title('Setosa 품종의 각 피쳐별 box plot')
plt.show()
```

```python
iris.boxplot(by='species', figsize=(10,8))
plt.show()
```
