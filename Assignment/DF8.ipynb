### 2. Titanic - sns.load_dataset('titanic')

<pre>
a. 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만드시오. 
category1 카테고리는 다음과 같이 정의됨

1) 20살이 넘으면 성별을 그대로 사용한다.

2) 20살 미만이면 성별에 관계없이 “child”라고 한다.
</pre>

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['category1'] = titanic.apply(lambda x: 'child' if x.age < 20 else x.sex, axis=1)
titanic.tail()
```

<pre>
b. 타이타닉호의 승객 중 나이를 명시하지 않은 고객은 
나이를 명시한 고객의 평균 나이 값이 되도록 titanic 데이터프레임을 고치시오.
</pre>

```python
titanic.age.fillna(titanic.age.mean(), inplace=True)
titanic.tail()
```

<pre>
c. 성별, 선실(class)별, 출발지(embark_town)별 생존율을 구하시오.
</pre>

```python
print(titanic.pivot_table('survived',['sex']))
print(titanic.pivot_table('survived',['class']))
print(titanic.pivot_table('survived',['embark_town']), end='')
```

```python
titanic.pivot_table('survived',['sex','class','embark_town']) * 100    # * 100은 %로 만들어줌.
```

<pre>
d. 타이타닉호 승객을 ‘미성년자’, ‘청년’, ‘중년’, ‘장년’, ‘노년’ 나이 그룹으로 나누고, 
각 그룹별 생존율을 구하시오.
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
</pre>

```python
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
cats = pd.cut(titanic.age, bins, labels=labels)
titanic.pivot_table('survived', cats)
```

e. qcut 명령으로 세 개의 나이 그룹을 만들고, 나이 그룹별 남녀 성비와 생존율을 구하시오.

```python
cats = pd.qcut(titanic.age, 3, labels='Q1 Q2 Q3'.split())
titanic.pivot_table('survived',[cats, 'sex']) * 100
```

### 3. Mile Per Gallon - sns.load_dataset('mpg')

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

<pre>
c. 엔진의 실린더(cylinders) 갯수별 연비(mpg)의 평균을 구하시오.
</pre>

```python
mpg.groupby('cylinders')[['mpg']].mean()
```

<pre>
d. 생산지(origin)별 배기량 대비 마력(hp_per_cc)의 평균을 구하시오.
</pre>

```python
mpg['hp_per_cc'] = mpg.horsepower / mpg.displacement
mpg.groupby('origin')[['hp_per_cc']].mean()
```

<pre>
e. 모델이 5개 이상인 제조사에 대하여
연비(mpg)의 평균이 가장 좋은 제조사 Top 5를 구하시오.
</pre>

```python
mpg['manufacturer'] = mpg.name.apply(lambda x: x.split()[0])
mpg['model'] = mpg.name.apply(lambda x: ' '.join(x.split()[1:]))
```

```python
mpg.manufacturer[mpg.manufacturer == 'chevrolet'].count()
```

```python
d = {}
mpg_manu = mpg.manufacturer
for column in mpg_manu.unique():
    if mpg_manu[mpg_manu == column].count() >= 5:
        d[column] = mpg.mpg[mpg_manu == column].mean()
sort_d = sorted(d.items(), key = lambda item: item[1], reverse=True)
df = pd.DataFrame(sort_d, columns=['manufacturer', 'mpg_mean'])
df
```

bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
그리고 각 나이 그룹의 승객 비율을 구한다. 비율의 전체 합은 1이 되어야 한다.
</pre>

```python
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
cats = pd.cut(titanic.age, bins, labels=labels)
for cat in cats.unique():
    d[cat] = (cats[cats == cat].count() / titanic.age.count()).round(2)
pd.DataFrame([d], index=['비율'])
```
