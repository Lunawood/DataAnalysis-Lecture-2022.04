# 데이터 시각화 - Seaborn 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.family'] = 'Malgun Gothic'

plt.rcParams.update({
                    'text.color': "gray",
                    'axes.labelcolor': "white",
                    'xtick.color': "white",     # xtick color
                    'ytick.color': "white"      # ytick color
                    })
```

### 1. Count plot

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.countplot(x='class', data=titanic)
plt.title('타이타닉호의 각 클래스별 승객수')
plt.show()
```

```python
sns.countplot(x='embark_town', data=titanic);
```

### Pair plot

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')           # 만약 카테고리형 데이터가 섞여 있는 경우에는 
                                            # hue 인수에 카테고리 변수 이름을 지정
plt.show()
```

### 3. Heatmap

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.shape
```

```python
flights.pivot('month', 'year', 'passengers')        # pivot: pivot_table의 열화판 연산x
```

```python
passengers = flights.pivot('month', 'year', 'passengers')
plt.figure(figsize=(10,8))
sns.heatmap(passengers, annot=True, fmt='d', linewidths=1)          # annot 글자를 쓸것인가?
plt.show()                                                          # fmt 데이터타입 d 정수
```

### 4. Box plot

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('요일별 요금')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips)
plt.title('요일별/성별 요금')
plt.show()
```

```python
sns.boxplot(data=iris[iris.species == 'setosa']);
```
