# 서울시 구별 CCTV/인구 분석

```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.family'] = 'Malgun Gothic'

plt.rcParams.update({'text.color': "white",
                    'axes.labelcolor': "white",
                    'xtick.color': "white",
                    'ytick.color': "white"
                    })
```

```python
df = pd.read_csv('../Project/cctv_pop.csv', index_col='구별')
df.head(3)
```

- 상관관계 분석
    - CCTV 댓수 vs 인구수, 외국인비율, 고령자비율

```python
# CCTV 댓수 vs 인구수
sns.regplot(x='인구수', y='CCTV댓수', data=df)
plt.show()      # 회귀선 regression
```

```python
_, axs = plt.subplots(1, 3, figsize=(15,5))
features = ['인구수', '외국인비율', '고령자비율']
for i, feature in enumerate(features):
    sns.regplot(x=feature, y='CCTV댓수', data=df, ax=axs[i])      # ax 내가 몇번째 그림을 그릴것인가
```

- 구별 CCTV 댓수 시각화

```python
df.CCTV댓수.plot(kind='barh')
plt.title('구별 CCTV 댓수')
plt.show()
```

```python
df.CCTV댓수.sort_values().plot(kind='barh', grid=True, figsize=(12, 6))
plt.title('구별 CCTV 댓수', fontsize=15)
plt.show()
```

- 인구대비 CCTV 비율 시각화

```python
# 인구 천명당 CCTV 댓수
df['CCTV비율'] = (df.CCTV댓수 / (df.인구수/1000)).round(2)
df.CCTV비율.sort_values().plot(kind='barh', grid=True, figsize=(12, 6))
plt.title('구별 인구 천명당 CCTV 댓수', fontsize=15)
plt.show()
```

### 인구수와 CCTV 댓수 분석
- 산점도(Scatter plot)

```python
plt.scatter(x=df.인구수, y=df.CCTV댓수)
plt.title('인구수와 CCTV 댓수의 산점도')
plt.xlabel('인구수'), plt.ylabel('CCTV 댓수')
plt.grid()
plt.show()
```

- 회귀선

```python
weight, bias = np.polyfit(df.인구수, df.CCTV댓수, 1)        # x, y, 차수(=>1,2,3... 차식)
weight, bias
```

```python
x = np.array([100000, 700000])
y = weight * x + bias
print(f'y: {y}')
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x=df.인구수, y=df.CCTV댓수)
plt.plot(x, y, 'g--', lw=3)                     # green(g), dashed(--)
plt.title('인구수와 CCTV 댓수의 산점도')
plt.xlabel('인구수'), plt.ylabel('CCTV 댓수')
plt.grid()
plt.show()
```

- 회귀선과 산점도 사이의 절대오차

```python
df['오차'] = abs(df.CCTV댓수 - (df.인구수 * weight + bias)).round(2)              # abs: 절댓값
df_sort = df.sort_values(by='오차', ascending=False)
df_sort.head(10)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x=df.인구수, y=df.CCTV댓수, c = df.오차, s = 50)        # c: color, s: dot size
plt.plot(x, y, 'g--', lw=3)                     
plt.title('인구수와 CCTV 댓수의 산점도')
plt.xlabel('인구수'), plt.ylabel('CCTV 댓수')
plt.colorbar()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(x=df.인구수, y=df.CCTV댓수, c = df.오차, s = 50)
plt.plot(x, y, 'g--', lw=3)                     
for i in range(8):
    plt.text(df_sort.인구수[i] + 9000, df_sort.CCTV댓수[i] - 50,
             df_sort.index[i], fontsize=15, color = 'black')
plt.title('인구수와 CCTV 댓수의 산점도')
plt.xlabel('인구수'), plt.ylabel('CCTV 댓수')
plt.colorbar()
plt.grid()
plt.show()
```
