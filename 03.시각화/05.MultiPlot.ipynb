# Matplotlib - 여려개의 그래프

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.family'] = 'Malgun Gothic'

plt.rcParams.update({'text.color': "white",
                    'axes.labelcolor': "white",
                    'xtick.color': "white",
                    'ytick.color': "white"
                    })
```

```python
from PIL import Image
img1 = Image.open('삼각함수.png')
img2 = Image.open('cat.jpg')

plt.figure(figsize = (8,4))
# 몇줄을 그릴것인가? 몇개의 컬럼을 그릴것인가? 몇번째 그림인가?
plt.subplot(121)    # 1줄에 2개의 컬럼 첫번째 사진
plt.imshow(img1)
plt.subplot(1, 2, 2)
plt.imshow(img2)
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t) 
def g(t):
    return np.sin(np.pi*t)

t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.01)
```

```python
plt.subplot(221)                # 2 x 2, 그래프중 첫번째
plt.plot(t1, f(t1))
plt.subplot(2,2,2)
plt.plot(t2, g(t2))
plt.subplot(223)
plt.plot(t1, f(t1), 'r-')
plt.subplot(2,2,4)
plt.plot(t2, g(t2), 'r-')
plt.show()
```

- Anscombe's Quartet

```python
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.head()
```

```python
ans.dataset.unique()
```

```python
ans.groupby('dataset').describe()
```

```python
# dataset I 을 선택해서 그려봄
ds = ans[ans.dataset == 'I']
plt.scatter(ds.x, ds.y, marker='o')
plt.plot([4, 14], [5, 10], 'r-', lw=2)      # y = 0.5 * x + 3
plt.title('Dataset I')
plt.show()
```

```python
plt.figure(figsize=(10, 8))

for index, roman in enumerate(ans.dataset.unique()):              # 0, 1, 2, 3 : ['I', 'II', 'III', 'IV']
    plt.subplot(2, 2, index + 1)
    ds = ans[ans.dataset == roman]
    plt.scatter(ds.x, ds.y, marker='o')
    plt.ylim(2, 14)
    plt.plot([2, 20], [4.5, 13], 'r-', lw=2)
    plt.title(f'Dataset {roman}')
plt.show()
```

- 좀더 세련된 그래프

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8), sharex =True, sharey = True)

for index, roman in enumerate(ans.dataset.unique()):              # 0, 1, 2, 3 : ['I', 'II', 'III', 'IV']
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.scatter(ds.x, ds.y, marker='o')
    ax.plot([2, 20], [4.5, 13], 'r-', lw=2)
    ax.set_title(f'Dataset {roman}')
fig.suptitle("Anscombe's Quartet", fontsize=16)
plt.tight_layout()
plt.show()
```
