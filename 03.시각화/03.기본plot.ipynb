# Matplotlib - 기본 plot

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['text.color'] = 'white'
```

```python
plt.plot([0,1,4,9])         # X, Y (X가 생략되면 0,1,2,3,... 을 의미)
plt.title('라인 플롯')
plt.grid()
plt.show()
```

- Line style

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ro-')        # color, marker(점표시), linestyle -> ro-: red-circle-직선
plt.title('Red-Circle-직선')
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--')
plt.plot([8,4,2,1], 'g^:')
plt.title('Magenta-Square-Dashed, Green-삼각형-점선')
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--', linewidth=3, markersize = 10)
plt.plot([8,4,2,1], 'g^:', lw=5, ms=20)
plt.title('Magenta-Square-Dashed, Green-삼각형-점선')
plt.grid()
plt.show()
```

```python
# Line style
# color: blue, green, red, cyan, megenta(보라색), yello, black(k), white
# marker: circle(o), 삼각현(^), 역삼각형(v), 사각형(s)
# line: line(-), dash(--)
```

- 그림 범위

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ro-')
plt.title('X축, Y축 범위 지정')
# 그림 범위
plt.xlim(-1, 4); plt.ylim(-1, 10)       # xlim: x_limit
plt.grid(); plt.show()
```

- 곡선 그림

```python
# 곡선 그림 함수
X = np.arange(0, 2*np.pi, 0.01)
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Sin, Cos 그래프')
plt.grid()
plt.show()
```

- Tick 설정

```python
X = np.arange(0, 2*np.pi, 0.01)
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Tick 설정')
# Tick 설정
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], 
            [r'$0$', r'$+\pi/2$', r'$+\pi$', r'$+3\pi/2$', r'$+2\pi$'])           # LaTex: pi를 문자로 보여줌.
plt.yticks([-1, 0, 1])
plt.grid()
plt.show()
```

- 범례(Legend)

```python
mpl.rcParams['text.color'] = 'green'
X = np.arange(0, 2*np.pi, 0.01)
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X, Ysin, label='sin'), plt.plot(X, Ycos, label='cos')
plt.title('범례 표시')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], 
            [r'$0$', r'$+\pi/2$', r'$+\pi$', r'$+3\pi/2$', r'$+2\pi$'], color='white')
plt.yticks([-1, 0, 1], color='white')
# 범례 표시
plt.legend(loc=0)
plt.grid()
plt.show()
```

- X축, Y축 Label

```python
# Text 생략

# 함수 생략

plt.plot(X, Ysin, label='sin'), plt.plot(X, Ycos, label='cos')
plt.title('X축, Y축 Label 표시', fontsize=15)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], 
            [r'$0$', r'$+\pi/2$', r'$+\pi$', r'$+3\pi/2$', r'$+2\pi$'], color='white')
plt.yticks([-1, 0, 1], color='white')
# X축, Y축 Label
plt.xlabel('시간', color='white'), plt.ylabel('진폭', color='white')
plt.legend(loc=0)
plt.grid()
plt.show()
```

- 그림의 크기

```python
# 함수 생략

# 그림 크기 지정
plt.figure(figsize=(10, 6))

plt.plot(X, Ysin, label='sin'), plt.plot(X, Ycos, label='cos')
plt.title('그림 크기 지정(10,6)', fontsize=15)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], 
            [r'$0$', r'$+\pi/2$', r'$+\pi$', r'$+3\pi/2$', r'$+2\pi$'], color='white')
plt.yticks([-1, 0, 1], color='white')
plt.xlabel('시간', color='white'), plt.ylabel('진폭', color='white')
plt.legend(loc=0)
plt.grid()
plt.show()
```

- 그림 저장

```python
# 함수 생략

# 그림 크기 지정
plt.figure(figsize=(10, 6))

plt.plot(X, Ysin, label='sin'), plt.plot(X, Ycos, label='cos')
plt.title('그림 저장', fontsize=15)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], 
            [r'$0$', r'$+\pi/2$', r'$+\pi$', r'$+3\pi/2$', r'$+2\pi$'], color='white')
plt.yticks([-1, 0, 1], color='white')
plt.xlabel('시간', color='white'), plt.ylabel('진폭', color='white')
plt.legend(loc=0)
plt.grid()
# 그림 저장
plt.savefig('삼각함수.png')
```
