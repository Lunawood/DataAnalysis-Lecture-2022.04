```python
import numpy as np
```

```python
x = np.arange(-10, 10, 1) # -10~9를 x에 대입
print(type(x))
print(x)
```

```python
y = 2*x - 1
print(y)
```

```python
# 그래프 그리기
import matplotlib.pyplot as plt
plt.rcParams.update({
    'text.color': "white",
    'axes.labelcolor': "white",

    'xtick.color': "white",
    'ytick.color': "white"
})

plt.plot(x, y)
plt.show()
```

```python
print("x 배열의 원소 개수: ", x.shape[0]) # x.shape[0] == len(x)
idx = np.arange(x.shape[0])  # np.arange(20) => 0~19 idx 대입
print("기존 인덱스: ", idx)
np.random.shuffle(idx)
print("섞인 인덱스: ", idx)
```

```python
x = x[idx]
print("x: ", x)
y = y[idx]
print("y: ", y)
```

```python
x_new = x.reshape(-1, 1)
print(x_new)
```

```python
# LinearRegression 클래스 함수는 통계학의 선형회귀분석을 처리.
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_new, y)    # 넣으므로써 lr에 식이 생김 y = 2*x-1
# 이때 x_new는 2차원 배열, y는 1차원 배열이다.
print("기울기: ", lr.coef_)
print("y절편: ", lr.intercept_)
```

```python
x_test = np.arange(11, 16, 1).reshape(-1 , 1)
x_test
```

```python
y_hat = lr.predict(x_test)  # 식을 응용 y=2*x-1
y_hat
```

```python
#
from tensorflow import keras
from tensorflow.keras.layers import Dense
```

```python
# Sequential 클래스 함수를 사용해서 신경만 객체를 생성
dnn = keras.Sequential()
```

```python
# units=1  즉 출력 데이터가 1가지 종류
# input_shape=(1,) 입력 데이터의 데이터 구조 지정
# 입력 변수 x가 1개라는 뜻...
dnn.add(Dense(units=1, input_shape=(1,)))
```

```python
# optimizer: 오차를 줄이는데 사용하는 최적화 알고리즘
# loss: 예측값의 오차를 측정하는 기준
dnn.compile(optimizer='sgd', loss='mse')
```

```python
# summary: 인공신경망 구조 출력
dnn.summary()
```

```python
# epochs=n n번 반복 학습
# y = 2*x-1
dnn.fit(x_new, y, epochs=1000)
```

```python
y_hat_dnn = dnn.predict(x_test)
y_hat_dnn
```
