```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# 데이터 파일 읽기
filename = 'cctv.csv'
cctv = pd.read_csv(filename, skiprows = 1, encoding='EUC-KR', thousands=',')
cctv.head()
```

```python
# Null data 있는지 확인하기
cctv.isna().sum().sum()
```

```python
# Null data는 0으로 채우기
cctv.fillna(0, inplace=True)
cctv.head()
```

```python
cctv.info()
```

```python
# 3. 1000 단위 구분기호 없애고 정수로 변환하기 thousand=','
cctv.info()
```

```python
cctv.head()
```

```python
# 2021년을 제외한 최근 3개년간의 CCTV 증가율 구하기
cctv['최근증가율'] = cctv[['2018년','2019년','2020년']].sum(axis=1) / cctv[['2012년 이전',
                    '2012년','2013년','2014년','2015년','2016년','2017년']].sum(axis=1) * 100
cctv.head()
```

```python
# 구분, 총계, 최근증가율 컬럼만 추출하기
inc_cctv = cctv['구분 총계 최근증가율'.split()]
inc_cctv.head()
```

```python
# 컬럼명을 구별, CCTV댓수로 변경하기
inc_cctv.rename(columns={'구분':'구별', '총계':'CCTV댓수'}, inplace=True)
inc_cctv.head()
```

```python
# 구 이름에서 공백지우기
inc_cctv['구별'] = inc_cctv['구별'].apply(lambda x: ''.join(x.split()))
inc_cctv.head()
```

```python
# 첫번째 행(계) 지우기
inc_cctv.drop(0, inplace=True)
inc_cctv.head()
```

<pre>
인구 데이터 정리하기
</pre>

```python
# 데이터 파일 읽기
filename = 'Seoul_population.txt'
pop = pd.read_csv(filename, skiprows = 2, sep='\t', thousands=',')
pop.head()
```

```python
# Null data 있는지 확인하기
pop.isna().sum().sum()
```

```python
# '자치구','계','계.1','계.2','65세이상고령자‘ 컬럼만 추출하고 이름을 '구별','인구수','내국인','외국인','고령자'로 변경
pop.rename(columns={'자치구':'구별','계':'인구수','계.1':'내국인','계.2':'외국인','65세이상고령자':'고령자'}, inplace=True)
pop.head()
```

```python
# 1000 단위 구분기호 없애고 정수로 변환하기 => thousand=','
pop.info()
```

```python
# 첫번째 행 지우기
pop.drop(0, inplace=True)
pop.head()
```

```python
# '외국인비율‘과 '고령자비율‘ 컬럼 만들기
pop['외국인비율'] = pop['외국인'] / pop['인구수']
pop['고령자비율'] = pop['고령자'] / pop['인구수']
pop.head()
```

```python
# 구 이름에 공백이 있으면 공백지우기
pop['구별'] = pop['구별'].apply(lambda x: ''.join(x.split()))
pop.head()
```

CCTV 데이터와 인구 데이터 합치고 저장하기
</pre>

```python
inc_cctv.head()
```

```python
pop.head()
```

```python
# 데이터 합치기
cctv_pop = pd.merge(inc_cctv, pop, on=['구별'])
cctv_pop.head()
```

```python
# ‘구별’ 컬럼을 인덱스로 만들기
cctv_pop.set_index('구별', inplace=True)
cctv_pop.head()
```

```python
# 합친 데이터를 cctv_pop.csv로 저장하기
cctv_pop.to_csv("cctv_pop.csv", mode='w')
```
