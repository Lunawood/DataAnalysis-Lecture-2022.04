# Pandas 시각화

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.family'] = 'Malgun Gothic'

plt.rcParams.update({'text.color': "gray",
                    'axes.labelcolor': "white",
                    'xtick.color': "white",
                    'ytick.color': "white"
                    })
```

### 1. Line graph

```python
np.random.seed(2022)
df = pd.DataFrame(np.random.randn(100, 3),                      # 3그룹으로 나누어서 100개씩 뽑겠다.
                    index = pd.date_range('2022-1-1', periods=100),
                    columns=['A', 'B', 'C']).cumsum()           # 누적합
df.head()
```

```python
df.plot()
plt.title('Pandas의 plot method 사용예시')
plt.xlabel('일자'), plt.ylabel('데이터')
plt.show()          # text color는 gray
```
